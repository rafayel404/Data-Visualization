```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#730385')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#073d69')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#730385',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#073d69',linestyle='dashed')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#730385',linestyle='solid')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#073d69',linestyle='dashdot')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python

plt.plot(batsman['index'],batsman['V Kohli'],color='#730385',linestyle='solid',linewidth=2.9,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#073d69',linestyle='dashdot',linewidth=1.2,marker='+',markersize=20)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#730385',linestyle='solid',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#073d69',linestyle='dashdot',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='lower right')
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='green')
```

```python
plt.barh(colors,children,color='green')
```

```python
# color and label
df = pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0])-.3,df['2015'],width=.3)
plt.bar(np.arange(df.shape[0]),df['2016'],width=.3)
plt.bar(np.arange(df.shape[0])+.3,df['2017'],width=.3)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
plt.bar(np.arange(df.shape[0])-.3,df['2015'],width=.3)
plt.bar(np.arange(df.shape[0]),df['2016'],width=.3)
plt.bar(np.arange(df.shape[0])+.3,df['2017'],width=.3)

plt.xticks(np.arange(df.shape[0]),df['batsman'],rotation=70)
plt.show()
```

```python
plt.bar(np.arange(df.shape[0])-.3,df['2015'],width=.3)
plt.bar(np.arange(df.shape[0]),df['2016'],width=.3)
plt.bar(np.arange(df.shape[0])+.3,df['2017'],width=.3)

plt.xticks(np.arange(df.shape[0]),df['batsman'],rotation=90)
plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%.2f%%')
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%.4f%%')
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%i%%')
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%.2f%%',explode=[.2,0,.2,0,0,0])
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%.2f%%',explode=[.2,0,.2,0,0,0],shadow=True,colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```
